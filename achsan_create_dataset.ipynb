```python
# create random seed
np.random.seed(76)

# Fungsi untuk menentukan lokasi berdasarkan tipe
def location(location_type):
    if location_type == 'Cafe':
        return np.random.choice(cafe_list, replace=True)
    elif location_type == 'Restaurant':
        return np.random.choice(restaurant_list, replace=True)
    elif location_type == 'Office':
        return np.random.choice(office_list, replace=True)
    elif location_type == 'Education Center':
        return np.random.choice(education_center, replace=True)
    elif location_type == 'Compound Space':
        return np.random.choice(compound_space, replace=True)
    elif location_type == 'Transportation':
        return np.random.choice(location_names, replace=True)
    else:
        return None  # Default jika tipe lokasi tidak cocok

# Terapkan fungsi ke dataframe
df['Location'] = df['Location Type'].apply(location)

# Tampilkan hasil
df.head(20)
```

## Generate Nomor Telefon

Menggunakan Wikipedia untuk mengetahui extensi awal dari sebuah provider di indonesia . seperti nomor2 dari 0852 0853 dansebagainya. [Sumber](https://id.wikipedia.org/wiki/Daftar_produk_telekomunikasi_di_Indonesia). 



```python
# create random seed
np.random.seed(96)

# list awalan nomor 
all_prefixes = [
    "0811", "0812", "0813", "0821", "0822", "0823", "0851", "0852", "0853",
    "0814", "0815", "0816", "0855", "0856", "0857", "0858",
    "0895", "0896", "0897", "0898", "0899",
    "0817", "0818", "0819", "0859", "0877", "0878",
    "0831", "0832", "0833", "0838",
    "0881", "0882", "0883", "0884", "0885", "0886", "0887", "0888", "0889"
]

# Fungsi untuk membuat nomor handphone acak
def generate_random_number():
    prefix = np.random.choice(all_prefixes)  # Pilih prefix acak
    suffix = ''.join(np.random.choice(list('0123456789'), size=8))  # 8 digit acak
    return prefix + suffix

# Tambahkan kolom Handphone ke DataFrame
df['Handphone'] = [generate_random_number() for _ in range(len(df))]

df.head()
```

## Generate Jam Login 

Kita tentukan dulu aja pertimbangan-pertimbangan jam login. Harusnya jam login manusia di jakarta untuk wifi kemungkinan dimulai dari jam 5 pagi sampai jam 10 malam. 5 pagi karena ada alasan orang-orang berangkat kantor dan kemudian untuk sampai jam 10 malam kita anggap itu orang terakhir keluar dari kantor ataupun mall di jakarta. 

```python
# create random seed
np.random.seed(555)
# jam login
jam_login = list(range(0, 25))

#probabilities untuk jam 
probabilities = [
    0.01, 0.01, 0.01, 0.01, 0.01, 0.01,  # 00:00 - 06:00 (Dini Hari)
    0.02, 0.02, 0.02,                   # 06:00 - 09:00 (Pagi Berangkat Kerja)
    0.05, 0.07, 0.10,                   # 09:00 - 12:00 (Aktivitas Pagi-Siang)
    0.15, 0.20,                         # 12:00 - 14:00 (Waktu Makan Siang)
    0.12, 0.10, 0.08, 0.07,             # 14:00 - 18:00 (Sore Nongkrong atau Belanja)
    0.15, 0.18, 0.20,                   # 18:00 - 21:00 (Malam Santai di Mall/Kafe)
    0.10, 0.08,                         # 21:00 - 22:00 (Persiapan Pulang)
    0.05, 0.03                          # 22:00 - 00:00 (Tengah Malam)
]  # asumsi penggunaan hotspot atau interaksi manusia misal jam berangkat kerja maka probabilitas rendah dan ketika makan siang tinggi dan sebagainya. 

# Buat Kolom Baru Login Hour
df['Login Hour'] = np.random.choice(jam_login,size=len(df))

# cek
df.head()
```

```python
# Inisialisasi Faker dan set seed
faker = Faker('id_ID')
random.seed(99)  # Seed untuk random
faker.seed_instance(99)  # Seed untuk Faker

# Daftar wilayah Jabodetabek
jabodetabek_areas = [
    "Jakarta Pusat", "Jakarta Barat", "Jakarta Selatan", 
    "Jakarta Timur", "Jakarta Utara", 
    "Bogor", "Tangerang", "Depok", "Bekasi","Tangerang Selatan"
]

def generate_jabodetabek_address():
    # Alamat dasar dari Faker + Wilayah Jabodetabek
    street_address = faker.street_address()
    area = random.choice(jabodetabek_areas)
    postcode = faker.postcode()
    return f"{street_address}, {area}, {postcode}"

# Buat DataFrame dengan kolom address
df['Address'] = [generate_jabodetabek_address() for _ in range(700)]  # Generate 10 alamat

# Tampilkan hasil
df.head()
```
